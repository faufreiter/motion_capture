```python
data_P1_ram = pd.read_csv('./training_P1_ram_final.csv')
#Documents/GitHub/motion_capture/data/P1/training_P1_ram_final.csv
#first few rows
#print(data_P1_ram.head())

#number of rows
num_rows_data_P1_ram = len(data_P1_ram)
print(f"rows: {num_rows_data_P1_ram}")

average_P1_ram = data_P1_ram.loc[470:3339,'P1_ram_final'].mean()
#frame info 

#print("Average P1_ram:", average_P1_ram)
average_P1_ram_FINAL = average_P1_ram/1000
average_P1_ram_FINAL
print(f"average_P1_ram_FINAL:{average_P1_ram_FINAL} angle")
```

```python
# Load the data from both CSV files
data_P1_lam = pd.read_csv('./training_P1_lam_final.csv')
data_P1_ram = pd.read_csv('./training_P1_ram_final.csv')

# Select the range of rows for plotting for both lam and ram
data_lam_to_plot = data_P1_lam.loc[3578:6500, 'angle_lam_final']
data_ram_to_plot = data_P1_ram.loc[470:3339, 'P1_ram_final']

# Create a line chart
plt.figure(figsize=(12, 8))

# Plotting lam data in red
plt.plot(data_lam_to_plot.index, data_lam_to_plot, label='angle_lam_final', color='red')

# Plotting ram data in blue
plt.plot(data_ram_to_plot.index, data_ram_to_plot, label='P1_ram_final', color='blue')

# Setting up the chart
plt.title('Overlay Line Chart of angle_lam_final and P1_ram_final')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()

# Show the plot
plt.show()
```

```python
# Load the data from both CSV files
data_P1_lam = pd.read_csv('./training_P1_lam_final.csv')
data_P1_ram = pd.read_csv('./training_P1_ram_final.csv')

# Select the range of rows for plotting for both lam and ram
data_lam_to_plot = data_P1_lam.loc[3578:6500, 'angle_lam_final'].reset_index(drop=True) / 1000
data_ram_to_plot = data_P1_ram.loc[470:3339, 'P1_ram_final'].reset_index(drop=True) / 1000

# Create a line chart
plt.figure(figsize=(12, 8))

# Plotting lam data in red
plt.plot(data_lam_to_plot.index / 100, data_lam_to_plot, label='P1: angle, left arm', color='red')

# Plotting ram data in blue
plt.plot(data_ram_to_plot.index / 100, data_ram_to_plot, label='P1: angle, right arm', color='blue')

# Setting up the chart
plt.title('Overlay Line Chart:\nP1:angle left & right arm')
plt.xlabel('Time (seconds)')
plt.ylabel('Angle (degrees)')

# Position the legend to the right of the title
plt.legend(loc='upper center', bbox_to_anchor=(0.7, 1), ncol=2)

# Show the plot
plt.show()
```

```python
data_P1_Rdist_traveled = pd.read_csv('./training_P1_Rdist_traveled.csv')
#Documents/GitHub/motion_capture/data/P1/training_P1_Rdist_traveled.csv
#first few rows
#print(data_P1_Rdist_traveled.head())

#number of rows
num_rows_data_P1_Rdist_traveled = len(data_P1_Rdist_traveled)
print(f"rows - data_P1_Rdist_traveled: {num_rows_data_P1_Rdist_traveled}")

max_value_R = data_P1_Rdist_traveled['right_elbow_Rdist_traveled'].max()
print(f"Maximum value, travelled distance: {max_value_R}")

specific_value = data_P1_Rdist_traveled['right_elbow_Rdist_traveled'][3339] - data_P1_Rdist_traveled['right_elbow_Rdist_traveled'][470]#like last value of biceps curls
print(f"Travelled distance, right elbow: {specific_value}")
```
